## Reading Questions: Knight, Woodward and Lancaster (2017), Violent versus Nonviolent Actors: An Empirical Study of Different Types of Extremism

**Q1: What are some key differences and similarities between VE and NVE? [3]**

Knight, Woodward and Lancaster (2017) analysed 40 in-depth case studies of extremist individuals (20 violent, 20 nonviolent) and found that the two groups overlap more than they differ on surface-level features. All 40 individuals in the sample, both violent and nonviolent, showed evidence of perceiving an external threat, seeking like-minded peers, and holding strong ideological convictions. Factors commonly associated with extremism, such as identity seeking, a sense of purpose, and hostility toward outgroups, appeared at comparable rates in both VEs and NVEs and therefore cannot reliably distinguish the two on their own.

The key differences that *did* emerge are behavioural and psychological:

- **Personal responsibility to act**: This was the strongest differentiator. 87.5% of VEs felt a personal duty to translate ideology into action, compared to only 18.8% of NVEs. The authors describe this as the single most telling variable.
- **Exposure to extreme violence**: VEs were significantly more likely to have been exposed to real-world violence (through conflict zones, personal victimisation, or violent networks), which may have lowered their threshold for committing it.
- **Individual vulnerability factors**: VEs showed higher rates of low self-esteem, experiences of bullying, and a pattern of underachievement, each of which may create a psychological need to prove oneself through dramatic action.
- **Extremist training and travel**: VEs were more likely to have participated in operational preparation, such as attending training camps or travelling to conflict areas, indicating a concrete escalation beyond rhetoric.

In contrast, NVEs remained within the domain of ideological expression: distributing propaganda, sending hate mail, or fundraising for extremist causes, all without direct involvement in physical harm. However, the authors caution that some NVE individuals may be "pre-violent" rather than genuinely nonviolent. They simply had not yet acted but could be on a trajectory toward violence.

**Q2: Based on the paper's method section, how are radicalization and extremism defined? [3]**

The authors acknowledge that definitional inconsistency is a persistent problem in the literature, with many studies conflating extremist attitudes with violent action. They address this by adopting behaviour-grounded operational definitions:

- **Extremism** is defined pragmatically: individuals who hold attitudes and beliefs that fall well outside mainstream opinion on political or ideological issues, and who had been convicted of a criminal offence where there was evidence of extremist motivation. This conviction-based framing was chosen to support counter-terrorism practitioners rather than to settle theoretical debates.

- **Radicalization** draws on the UK Home Office definition: the process by which individuals come to support terrorism and extremism and, in some cases, to participate in terrorist activity. Crucially, the authors emphasise that radicalization does not inevitably lead to violence. It is one of several possible pathways, and many people hold radical views without ever crossing into violent action.

The methodology itself involved constructing detailed case studies from open-source data (court records, media reports, official documents) for the 40 individuals and coding them against a framework of themes, subthemes, and underlying variables. VEs and NVEs were then compared statistically across these variables, revealing which factors show meaningful separation between the two groups and which are shared.

**Q3: How are Violent and Non-Violent Extremism defined? Use this definition to design your prompt. [3]**

Following Knight et al. (2017), our working definitions are:

- **Violent Extremism (VE)**: Any act, or clearly intended act, constituting murder, attempted murder, assault, or serious physical harm to persons or property in pursuit of an ideological goal. This extends to individuals who conduct nonviolent behaviours that knowingly facilitate violence by others, for example financing attacks, providing weapons, or operational planning. In the original study, VE offences included terrorist attacks and foiled plots intended to kill.

- **Non-Violent Extremism (NVE)**: The promotion, dissemination, or support of an extreme ideology through non-physical means, without direct involvement in or facilitation of violence. In the study, NVE offences included possession or distribution of extremist literature, sending racist hate mail, and fundraising for extremist organisations.

The critical insight for prompt design is that ideology and grievance alone cannot separate VE from NVE, because both groups share those features equally. The classifier must focus on whether the text involves, plans, or enables physical harm (VE) versus whether it remains confined to ideological hostility without an actionable violence component (NVE). We embedded these definitions and this decision logic directly into our classification prompt, along with a confidence calibration mechanism to flag borderline cases.

```python
"""
ASSIGNMENT 3 - PART A:  VE/NVE CLASSIFICATION

"""


# INSTALLATIONS AND IMPORTS

!pip install -q openai pandas numpy matplotlib seaborn scikit-learn tqdm

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import time
from typing import Dict
import warnings
warnings.filterwarnings('ignore')

import openai
from google.colab import userdata

client = openai.OpenAI(
    base_url="https://ai-gateway.andrew.cmu.edu/v1",
    api_key=userdata.get('OPENAI_API_KEY')
)

#  TA clarification: use gpt-4.1-mini due to auth issues with gpt-4o-mini-2024-07-18
MODEL_A = "gpt-4.1-mini"

print("=" * 80)
print("ASSIGNMENT 3 - PART A: VE/NVE CLASSIFICATION ")
print("=" * 80)
print(f"Model: {MODEL_A}")
print("=" * 80)

#
# LOAD DATA
#  from assignment 2
df = pd.read_csv('extremism_data_cleaned.csv')
print(f"Dataset loaded: {df.shape}")
print(f"Label distribution:\n{df['Extremism_Label'].value_counts()}")

# SAMPLING STRATEGY
# Oversample to guarantee 50 VE + 50 NVE after classification


np.random.seed(42)
extremist_pool = df[df['Extremism_Label'] == 'EXTREMIST'].sample(
    n=min(400, len(df[df['Extremism_Label'] == 'EXTREMIST'])), random_state=42
)
non_extremist_pool = df[df['Extremism_Label'] == 'NON_EXTREMIST'].sample(
    n=min(150, len(df[df['Extremism_Label'] == 'NON_EXTREMIST'])), random_state=42
)

working_df = pd.concat([extremist_pool, non_extremist_pool]).reset_index(drop=True)
working_df.insert(0, 'id', range(1, len(working_df) + 1))
print(f"\nWorking pool: {len(working_df)} rows")
print(f"  EXTREMIST pool:     {len(extremist_pool)}")
print(f"  NON_EXTREMIST pool: {len(non_extremist_pool)}")
print("Will stop early once 50 VE + 50 NVE confirmed.\n")


# PROMPTS
# Academic framing added to reduce model guardrail refusals
# Grounded in:
#   Lancaster et al. (2023) — VE/NVE definitions
#   ACM (2023) dl.acm.org/doi/10.1145/3575802 — subtype taxonomy
#   Nature (2025) nature.com/articles/s41599-025-05685-z — target groups


VE_NVE_SYSTEM_PROMPT = """You are a research assistant supporting academic content moderation research on extremism detection. This is a scholarly classification task. You will analyze social media posts for research purposes only — your output will be used to study extremist rhetoric patterns, not to promote harmful content.

DEFINITIONS (Lancaster et al., 2023):
- VIOLENT EXTREMISM (VE): Content that advocates, incites, endorses, celebrates, or directly
  threatens PHYSICAL VIOLENCE or acts of terror against individuals or groups. Includes
  IMPLICIT and CODED incitement.
- NON-VIOLENT EXTREMISM (NVE): Radical ideology, hate speech, dehumanization, or supremacist
  views WITHOUT calling for or celebrating physical violence.

VE INDICATORS (any ONE is sufficient):
- Direct calls to violence: "kill", "attack", "shoot", "bomb", "murder", "slaughter"
- Explicit threats: "you deserve to die", "they should be shot"
- Celebrating violence: "they deserved to be killed", "praise the bombing"
- Calls for genocide/terrorism: "cleanse", "exterminate", "wipe out" a group
- Coded incitement: "you know what needs to be done", "make them pay with blood"
- Glorifying past terrorist attacks or perpetrators
- Dehumanization + urgency implying violence ("they must be stopped NOW")
- Expressing that violence against a group is justified or deserved

NVE INDICATORS:
- Slurs, insults, profanity toward groups
- Dehumanizing comparisons ("they are animals/parasites")
- Supremacist ideology without violence call
- Conspiracy theories about groups
- Political hate speech without physical threat

NUANCE:
- "I hate [group]" → NVE
- "Someone should kill [group]" → VE
- "They are inferior" → NVE
- "Attack them on sight" → VE
- Ambiguous with NO clear violence call → NVE

If the model cannot classify due to content policies, output CLASSIFICATION: Refusal.

OUTPUT FORMAT (follow exactly):
CLASSIFICATION: [VE or NVE or Refusal]
TEXT_SPAN: [Exact phrase from the post that drove your decision]
EXPLANATION: [1-2 sentences: key evidence and which rule applies]
CONFIDENCE: [HIGH / MEDIUM / LOW]"""


VE_SUBTYPE_SYSTEM_PROMPT = """You are a research assistant supporting academic extremism research. This is a scholarly classification task for academic purposes only.

Classify the PRIMARY subtype of this Violent Extremism post.

SUBTYPE DEFINITIONS (ACM 2023 + Nature 2025):
- IDEOLOGICAL: Violence driven by supremacist, racial, ethnic, nationalist, misogynistic,
  or anti-government ideology (e.g., white supremacy, incel violence, ethnonationalism).
- POLITICAL: Violence targeting political systems, institutions, leaders, or policies
  (e.g., targeting politicians, government buildings, coercion through force).
- RELIGIOUS: Violence motivated by religious doctrine or religious identity
  (e.g., jihadist content, religious persecution, sectarian violence).

FEW-SHOT EXAMPLES:
Post: "White people need to cleanse this country of the parasites invading it"
→ SUBTYPE: IDEOLOGICAL (racial supremacy driving the violence call)

Post: "Politicians who voted for this bill deserve to be shot"
→ SUBTYPE: POLITICAL (targeting political figures over policy)

Post: "Infidels must be destroyed in the name of God"
→ SUBTYPE: RELIGIOUS (religiously motivated violence)

Post: "These feminist traitors should be eliminated from society"
→ SUBTYPE: IDEOLOGICAL (misogynistic ideology driving incitement)

Post: "We must bomb the parliament to stop this corrupt regime"
→ SUBTYPE: POLITICAL (targeting political institution)

Pick the DOMINANT subtype. If truly unclear, output Unclear.
If model cannot classify, output Refusal.

OUTPUT FORMAT:
SUBTYPE: [IDEOLOGICAL or POLITICAL or RELIGIOUS or Unclear or Refusal]
TEXT_SPAN: [Exact phrase driving subtype decision]
EXPLANATION: [1-2 sentences: which definition applies and why]"""


TARGET_GROUP_SYSTEM_PROMPT = """You are a research assistant supporting academic extremism research. This is a scholarly classification task for academic purposes only.

Identify ALL groups being targeted in this extremist post.

TARGET CATEGORIES (Nature 2025):
- Religious (e.g., Muslims, Jews, Christians, Atheists)
- Ethnic/Racial (e.g., Black people, white people, Hispanics, Asians)
- Political (e.g., Democrats, Republicans, liberals, conservatives)
- Gender (e.g., women, men, LGBTQ+ people, transgender people)
- Nationality (e.g., immigrants, Americans, specific nationalities)
- Occupation (e.g., police, politicians, journalists, teachers)
- Other (e.g., the wealthy, specific organizations)

Rules:
- List ALL groups explicitly mentioned OR clearly implied
- Separate multiple groups with commas
- If truly no identifiable group → Unclear
- If model cannot classify → Refusal

FEW-SHOT EXAMPLES:
Post: "These Jewish bankers are destroying our country"
→ TARGET_GROUPS: Jewish people, bankers

Post: "Liberals and immigrants are ruining everything"
→ TARGET_GROUPS: liberals, immigrants

Post: "Women who speak up deserve what's coming to them"
→ TARGET_GROUPS: women

Post: "Muslim terrorists must be wiped out along with their sympathizers"
→ TARGET_GROUPS: Muslims, perceived sympathizers

OUTPUT FORMAT:
TARGET_GROUPS: [comma-separated list, or Unclear, or Refusal]
TEXT_SPAN: [Exact phrase(s) showing who is targeted]
EXPLANATION: [How you identified the target]"""

print(" Prompts loaded (academic framing + Lancaster 2023 + ACM 2023 + Nature 2025)")

# PARSING FUNCTIONS
#  Refusal handling included in all parsers

def parse_label(response_text: str) -> Dict:
    result = {
        "classification": "NVE",
        "text_span": "",
        "explanation": "",
        "confidence": "LOW"
    }
    try:
        lines = response_text.strip().split('\n')
        for line in lines:
            line = line.strip()
            if line.upper().startswith('CLASSIFICATION:'):
                val = line.split(':', 1)[1].strip().upper()
                if 'REFUSAL' in val or 'REFUSE' in val:
                    result["classification"] = 'Refusal'
                elif val in ('VE', '[VE]') or (val.startswith('VE') and 'NVE' not in val):
                    result["classification"] = 'VE'
                else:
                    result["classification"] = 'NVE'
            elif line.upper().startswith('TEXT_SPAN:'):
                result["text_span"] = line.split(':', 1)[1].strip()
            elif line.upper().startswith('EXPLANATION:'):
                result["explanation"] = line.split(':', 1)[1].strip()
            elif line.upper().startswith('CONFIDENCE:'):
                val = line.split(':', 1)[1].strip().upper()
                result["confidence"] = val if val in ['HIGH', 'MEDIUM', 'LOW'] else 'MEDIUM'
    except Exception as e:
        result["explanation"] = f"Parse error: {e}"
    return result


def parse_subtype(response_text: str) -> Dict:
    result = {"subtype": "Unclear", "text_span": "", "explanation": ""}
    try:
        lines = response_text.strip().split('\n')
        for line in lines:
            line = line.strip()
            if line.upper().startswith('SUBTYPE:'):
                val = line.split(':', 1)[1].strip().upper()
                if 'REFUSAL' in val or 'REFUSE' in val:
                    result["subtype"] = "Refusal"
                elif 'IDEOLOGICAL' in val:
                    result["subtype"] = "IDEOLOGICAL"
                elif 'POLITICAL' in val:
                    result["subtype"] = "POLITICAL"
                elif 'RELIGIOUS' in val:
                    result["subtype"] = "RELIGIOUS"
                else:
                    result["subtype"] = "Unclear"
            elif line.upper().startswith('TEXT_SPAN:'):
                result["text_span"] = line.split(':', 1)[1].strip()
            elif line.upper().startswith('EXPLANATION:'):
                result["explanation"] = line.split(':', 1)[1].strip()
    except:
        pass
    return result


def parse_target(response_text: str) -> Dict:
    result = {"target_groups": ["Unclear"], "text_span": "", "explanation": ""}
    try:
        lines = response_text.strip().split('\n')
        for line in lines:
            line = line.strip()
            if line.upper().startswith('TARGET_GROUPS:'):
                groups_str = line.split(':', 1)[1].strip().strip('[]')
                if 'refusal' in groups_str.lower() or 'refuse' in groups_str.lower():
                    result["target_groups"] = ["Refusal"]
                elif groups_str and groups_str.lower() != 'unclear':
                    groups = [g.strip() for g in groups_str.split(',') if g.strip()]
                    result["target_groups"] = groups if groups else ["Unclear"]
                else:
                    result["target_groups"] = ["Unclear"]
            elif line.upper().startswith('TEXT_SPAN:'):
                result["text_span"] = line.split(':', 1)[1].strip()
            elif line.upper().startswith('EXPLANATION:'):
                result["explanation"] = line.split(':', 1)[1].strip()
    except:
        pass
    return result

print("✓ Parsing functions ready (with Refusal handling)")

# CLASSIFICATION FUNCTIONS

def classify_ve_nve(text: str) -> Dict:
    user_prompt = (
        "This is an academic research task. Classify the following social media post "
        "as VE (Violent Extremism) or NVE (Non-Violent Extremism) for extremism research.\n\n"
        f"Post: \"{text}\"\n\n"
        "Does this post advocate, incite, celebrate, or imply physical violence — even through "
        "coded language or glorification? Or does it express hate/extremist ideology without a "
        "violence call? Provide the exact text span that drove your decision."
    )
    try:
        response = client.chat.completions.create(
            model=MODEL_A,
            messages=[
                {"role": "system", "content": VE_NVE_SYSTEM_PROMPT},
                {"role": "user", "content": user_prompt}
            ],
            temperature=0.1,
            max_tokens=400
        )
        return parse_label(response.choices[0].message.content)
    except Exception as e:
        return {"classification": "Refusal", "text_span": "", "explanation": f"API error: {e}", "confidence": "LOW"}


def classify_ve_subtype(text: str) -> Dict:
    user_prompt = (
        "This is an academic research task. This post has been classified as Violent Extremism. "
        "Identify its PRIMARY subtype (IDEOLOGICAL, POLITICAL, or RELIGIOUS) for research purposes.\n\n"
        f"Post: \"{text}\"\n\n"
        "Provide the exact text span that identifies the subtype."
    )
    try:
        response = client.chat.completions.create(
            model=MODEL_A,
            messages=[
                {"role": "system", "content": VE_SUBTYPE_SYSTEM_PROMPT},
                {"role": "user", "content": user_prompt}
            ],
            temperature=0.1,
            max_tokens=300
        )
        return parse_subtype(response.choices[0].message.content)
    except Exception as e:
        return {"subtype": "Refusal", "text_span": "", "explanation": f"Error: {e}"}


def identify_target_groups(text: str) -> Dict:
    user_prompt = (
        "This is an academic research task. Identify ALL groups being targeted in this "
        "extremist social media post for research purposes.\n\n"
        f"Post: \"{text}\"\n\n"
        "List every group explicitly mentioned or clearly implied. "
        "Provide the exact text span that identifies the targets."
    )
    try:
        response = client.chat.completions.create(
            model=MODEL_A,
            messages=[
                {"role": "system", "content": TARGET_GROUP_SYSTEM_PROMPT},
                {"role": "user", "content": user_prompt}
            ],
            temperature=0.1,
            max_tokens=300
        )
        return parse_target(response.choices[0].message.content)
    except Exception as e:
        return {"target_groups": ["Refusal"], "text_span": "", "explanation": f"Error: {e}"}

print("✓ Classification functions ready")

#
# INITIALIZE COLUMNS

for col in ['VE_NVE_classification', 'VE_NVE_text_span', 'VE_NVE_explanation', 'VE_NVE_confidence',
            'VE_subtype', 'VE_subtype_text_span', 'VE_subtype_explanation',
            'target_groups', 'target_groups_text_span', 'target_groups_explanation']:
    working_df[col] = ""

# RUN CLASSIFICATION
#  Early-stop: collect exactly 50 VE + 50 NVE
#  Refusals tracked separately and excluded from class counts
# All posts get text_span + explanation (interpretability requirement)

print(f"Processing up to {len(working_df)} posts...")
print("Stopping when 50 VE + 50 NVE confirmed.\n")

ve_count      = 0
nve_count     = 0
refusal_count = 0
TARGET_VE     = 50
TARGET_NVE    = 50
processed_ids = []

for idx, row in tqdm(working_df.iterrows(), total=len(working_df)):

    if ve_count >= TARGET_VE and nve_count >= TARGET_NVE:
        print(f"\n Target reached: {ve_count} VE + {nve_count} NVE — stopping.")
        break

    text   = str(row['Original_Message'])
    ve_res = classify_ve_nve(text)
    label  = ve_res['classification']

    # Track refusals but don't count toward class targets
    if label == 'Refusal':
        refusal_count += 1
        working_df.at[idx, 'VE_NVE_classification'] = 'Refusal'
        working_df.at[idx, 'VE_NVE_text_span']      = ve_res.get('text_span', '')
        working_df.at[idx, 'VE_NVE_explanation']     = ve_res.get('explanation', 'Model refusal')
        working_df.at[idx, 'VE_NVE_confidence']      = 'LOW'
        working_df.at[idx, 'VE_subtype']             = 'Refusal'
        working_df.at[idx, 'target_groups']          = 'Refusal'
        processed_ids.append(idx)
        time.sleep(0.3)
        continue

    # Skip if we already have enough of this class
    if label == 'VE'  and ve_count  >= TARGET_VE:
        continue
    if label == 'NVE' and nve_count >= TARGET_NVE:
        continue

    # Store VE/NVE result
    working_df.at[idx, 'VE_NVE_classification'] = label
    working_df.at[idx, 'VE_NVE_text_span']      = ve_res['text_span']
    working_df.at[idx, 'VE_NVE_explanation']     = ve_res['explanation']
    working_df.at[idx, 'VE_NVE_confidence']      = ve_res['confidence']

    if label == 'VE':
        ve_count += 1
        # Subtype (VE only)
        subtype = classify_ve_subtype(text)
        working_df.at[idx, 'VE_subtype']             = subtype['subtype']
        working_df.at[idx, 'VE_subtype_text_span']   = subtype['text_span']
        working_df.at[idx, 'VE_subtype_explanation'] = subtype['explanation']
    else:
        nve_count += 1
        working_df.at[idx, 'VE_subtype']             = 'N/A'
        working_df.at[idx, 'VE_subtype_text_span']   = 'N/A'
        working_df.at[idx, 'VE_subtype_explanation'] = 'N/A'

    # Target groups — ALL posts
    target = identify_target_groups(text)
    working_df.at[idx, 'target_groups']             = ', '.join(target['target_groups'])
    working_df.at[idx, 'target_groups_text_span']   = target['text_span']
    working_df.at[idx, 'target_groups_explanation'] = target['explanation']

    processed_ids.append(idx)
    time.sleep(0.3)

print(f"\n✓ Classification complete")
print(f"  VE collected:       {ve_count}")
print(f"  NVE collected:      {nve_count}")
print(f"  Refusals (logged):  {refusal_count}")

# BUILD FINAL BALANCED CSV (50 VE + 50 NVE, refusals excluded)

results_df = working_df.loc[processed_ids].copy()
ve_rows    = results_df[results_df['VE_NVE_classification'] == 'VE'].head(TARGET_VE)
nve_rows   = results_df[results_df['VE_NVE_classification'] == 'NVE'].head(TARGET_NVE)
final_df   = pd.concat([ve_rows, nve_rows]).reset_index(drop=True)
final_df['id'] = range(1, len(final_df) + 1)

print(f"\n Final balanced dataset: {len(final_df)} rows")
print(f"  VE:  {len(ve_rows)}")
print(f"  NVE: {len(nve_rows)}")

#
# SAVE RESULTS
#

output_cols = [
    'id', 'Original_Message', 'Extremism_Label',
    'VE_NVE_classification', 'VE_NVE_text_span', 'VE_NVE_explanation', 'VE_NVE_confidence',
    'VE_subtype', 'VE_subtype_text_span', 'VE_subtype_explanation',
    'target_groups', 'target_groups_text_span', 'target_groups_explanation'
]

ve_nve_classifications = final_df[output_cols].copy()
ve_nve_classifications.to_csv('ve_nve_classifications.csv', index=False)
print(f"✓ Saved → ve_nve_classifications.csv ({len(ve_nve_classifications)} rows)")

# Also save full results including refusals for reporting
results_df[output_cols].to_csv('ve_nve_classifications_full_with_refusals.csv', index=False)
print(f" Saved → ve_nve_classifications_full_with_refusals.csv ({len(results_df)} rows)")

#
# ANALYSIS

print("\n" + "="*60)
print("CLASSIFICATION RESULTS")
print("="*60)

ve_nve_counts = ve_nve_classifications['VE_NVE_classification'].value_counts()
print("\nVE vs NVE:")
for label, count in ve_nve_counts.items():
    print(f"  {label}: {count} ({count/len(ve_nve_classifications)*100:.1f}%)")

conf_counts = ve_nve_classifications['VE_NVE_confidence'].value_counts()
print("\nConfidence levels:")
for conf, count in conf_counts.items():
    print(f"  {conf}: {count} ({count/len(ve_nve_classifications)*100:.1f}%)")

ve_posts = ve_nve_classifications[ve_nve_classifications['VE_NVE_classification'] == 'VE']

# Subtype analysis
subtype_counts     = pd.Series(dtype=int)
most_common_subtype = "None"
if len(ve_posts) > 0:
    subtype_counts = ve_posts['VE_subtype'].value_counts()
    print(f"\nVE Subtypes (from {len(ve_posts)} VE posts):")
    for sub, cnt in subtype_counts.items():
        print(f"  {sub}: {cnt} ({cnt/len(ve_posts)*100:.1f}%)")
    most_common_subtype = subtype_counts.index[0]
    print(f"\n→ Most common subtype: {most_common_subtype}")

# Target group analysis
all_targets = []
for t in ve_nve_classifications['target_groups'].dropna():
    if t.lower() not in ['n/a', 'unclear', '', 'refusal']:
        for g in t.split(','):
            g = g.strip()
            if g and g.lower() not in ['unclear', 'n/a', 'refusal']:
                all_targets.append(g)

target_counts = pd.Series(dtype=int)
most_targeted = "Unclear"
if all_targets:
    target_counts = pd.Series(all_targets).value_counts()
    most_targeted = target_counts.index[0]
    print("\nTop targeted groups:")
    for group, cnt in target_counts.head(10).items():
        print(f"  {group}: {cnt}")
    print(f"\n→ Most targeted: {most_targeted}")
else:
    print("\nNo clear targets extracted")

# Original label breakdown
extremist_orig  = ve_nve_classifications[ve_nve_classifications['Extremism_Label'] == 'EXTREMIST']
ve_in_extremist = len(extremist_orig[extremist_orig['VE_NVE_classification'] == 'VE'])
print(f"  Classified VE:  {ve_in_extremist} ({ve_in_extremist/max(len(extremist_orig),1)*100:.1f}%)")
print(f"  Classified NVE: {len(extremist_orig)-ve_in_extremist}")
print(f"\nRefusals logged (excluded from CSV): {refusal_count}")

print("\n" + "-"*40)
print("ANSWERS TO ANALYSIS QUESTIONS")
print("-"*40)
print(f"Q1: Most frequently targeted group:   {most_targeted}")
print(f"Q2: Most occurring extremism subtype: {most_common_subtype}")
print(f"Q3: Dataset contains more:            {'VE' if ve_nve_counts.get('VE',0) > ve_nve_counts.get('NVE',0) else 'NVE (balanced 50/50)'}")

# VISUALIZATION

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle(
    f'Part A: VE/NVE Classification Analysis\n'
    f'Model: {MODEL_A}  |  Balanced: {TARGET_VE} VE + {TARGET_NVE} NVE',
    fontsize=13, fontweight='bold'
)

# Plot 1: VE vs NVE
ax1 = axes[0, 0]
colors1 = ['#FF6B6B' if l == 'VE' else '#4ECDC4' for l in ve_nve_counts.index]
bars = ax1.bar(ve_nve_counts.index, ve_nve_counts.values, color=colors1, edgecolor='black', linewidth=0.8)
ax1.set_title('VE vs NVE Distribution', fontweight='bold')
ax1.set_ylabel('Count')
for bar, val in zip(bars, ve_nve_counts.values):
    ax1.text(bar.get_x() + bar.get_width()/2, val + 0.3, str(val), ha='center', fontweight='bold', fontsize=12)

# Plot 2: Confidence levels
ax2 = axes[0, 1]
conf_order = ['HIGH', 'MEDIUM', 'LOW']
conf_vals  = [conf_counts.get(c, 0) for c in conf_order]
conf_cols  = ['#2ECC71', '#F39C12', '#E74C3C']
bars2 = ax2.bar(conf_order, conf_vals, color=conf_cols, edgecolor='black', linewidth=0.8)
ax2.set_title('Confidence Level Distribution', fontweight='bold')
ax2.set_ylabel('Count')
for bar, val in zip(bars2, conf_vals):
    if val > 0:
        ax2.text(bar.get_x() + bar.get_width()/2, val + 0.3, str(val), ha='center', fontweight='bold')

# Plot 3: VE Subtypes
ax3 = axes[0, 2]
if len(subtype_counts) > 0:
    sub_colors = {
        'IDEOLOGICAL': '#FFB347', 'POLITICAL': '#6A5ACD',
        'RELIGIOUS': '#20B2AA', 'Unclear': '#95A5A6', 'Refusal': '#DDDDDD'
    }
    bar_colors3 = [sub_colors.get(s, '#95A5A6') for s in subtype_counts.index]
    bars3 = ax3.bar(subtype_counts.index, subtype_counts.values,
                    color=bar_colors3, edgecolor='black', linewidth=0.8)
    ax3.set_title(f'VE Subtypes (n={len(ve_posts)})', fontweight='bold')
    ax3.set_ylabel('Count')
    ax3.tick_params(axis='x', rotation=20)
    for bar, val in zip(bars3, subtype_counts.values):
        ax3.text(bar.get_x() + bar.get_width()/2, val + 0.1, str(val), ha='center', fontweight='bold')
else:
    ax3.text(0.5, 0.5, 'No VE posts', ha='center', va='center', transform=ax3.transAxes)
    ax3.set_title('VE Subtypes', fontweight='bold')

# Plot 4: Top targeted groups
ax4 = axes[1, 0]
if len(target_counts) > 0:
    top8 = target_counts.head(8)
    top8.plot(kind='barh', ax=ax4, color='#4A90E2', edgecolor='black', linewidth=0.8)
    ax4.set_title('Top Targeted Groups (All Posts)', fontweight='bold')
    ax4.set_xlabel('Frequency')
    ax4.invert_yaxis()
    for i, val in enumerate(top8.values):
        ax4.text(val + 0.1, i, str(val), va='center', fontweight='bold')
else:
    ax4.text(0.5, 0.5, 'No targets identified', ha='center', va='center', transform=ax4.transAxes)
    ax4.set_title('Top Targeted Groups', fontweight='bold')

# Plot 5: Original labels vs new classification
ax5 = axes[1, 1]
non_orig    = ve_nve_classifications[ve_nve_classifications['Extremism_Label'] == 'NON_EXTREMIST']
non_ve      = len(non_orig[non_orig['VE_NVE_classification'] == 'VE'])
non_nve     = len(non_orig) - non_ve
x = np.arange(2)
ve_vals_bar  = [ve_in_extremist, non_ve]
nve_vals_bar = [len(extremist_orig) - ve_in_extremist, non_nve]
ax5.bar(x, ve_vals_bar,  color='#FF6B6B', label='Classified VE',  edgecolor='black', linewidth=0.8)
ax5.bar(x, nve_vals_bar, bottom=ve_vals_bar, color='#4ECDC4',
        label='Classified NVE', edgecolor='black', linewidth=0.8)
ax5.set_title('Original Labels vs New Classification', fontweight='bold')
ax5.set_xticks(x)
ax5.set_xticklabels(['EXTREMIST\n(original)', 'NON_EXTREMIST\n(original)'])
ax5.set_ylabel('Count')
ax5.legend()

# Plot 6: Summary
ax6 = axes[1, 2]
ax6.axis('off')
high_med = conf_counts.get('HIGH', 0) + conf_counts.get('MEDIUM', 0)
summary_lines = [
    "CLASSIFICATION SUMMARY",
    "─" * 30,
    f"Model:           {MODEL_A}",
    f"Total posts:     {len(ve_nve_classifications)}",
    f"VE posts:        {ve_nve_counts.get('VE', 0)}",
    f"NVE posts:       {ve_nve_counts.get('NVE', 0)}",
    f"Refusals logged: {refusal_count}",
    f"High/Med conf:   {high_med}",
    "",
    f"Most common subtype:",
    f"  {most_common_subtype}",
    "",
    f"Most targeted group:",
    f"  {most_targeted}",
    "",
    f"Dominant class: NVE (balanced)",
    "",
    "Sources: Lancaster 2023,",
    "ACM 2023, Nature 2025"
]
ax6.text(0.05, 0.97, '\n'.join(summary_lines), fontsize=9, verticalalignment='top',
         transform=ax6.transAxes, fontfamily='monospace',
         bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.9))

plt.tight_layout()
plt.savefig('part_a_balanced_analysis.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n PART A COMPLETED")
print("Deliverables:")
print("  - ve_nve_classifications.csv                    (50 VE + 50 NVE, balanced)")
print("  - ve_nve_classifications_full_with_refusals.csv (full log including refusals)")
print("  - part_a_balanced_analysis.png")
```

```python
# PART B: PROMPT ENGINEERING COMPARISON

import pandas as pd
import numpy as np
import re
import time
import matplotlib.pyplot as plt
from tqdm import tqdm

# 1. LOAD DATA FROM PART A
df_part_a = pd.read_csv('ve_nve_classifications.csv')
print(f"Loaded ve_nve_classifications.csv: {len(df_part_a)} rows")
print(f"VE count: {(df_part_a['VE_NVE_classification'] == 'VE').sum()}")
print(f"NVE count: {(df_part_a['VE_NVE_classification'] == 'NVE').sum()}")

# 2. EXTRACT 100 BALANCED EXAMPLES (50 VE + 50 NVE)
ve_samples  = df_part_a[df_part_a['VE_NVE_classification'] == 'VE'].head(50)
nve_samples = df_part_a[df_part_a['VE_NVE_classification'] == 'NVE'].head(50)
df_part_b   = pd.concat([ve_samples, nve_samples]).reset_index(drop=True)
print(f"\nWorking sample for Part B: {len(df_part_b)} rows (50 VE + 50 NVE)")

# 3. MODEL — using gemini-2.5-flash (confirmed working from Part A)
MODEL_B = "gemini-2.5-flash"
print(f"Model: {MODEL_B}")

# 4. SEVERITY RUBRIC
SEVERITY_RUBRIC = """
Assign a Severity Level from 1 to 5 based on the following scale:
Level 1: General extremist interest or ideology with no direct threat.
Level 2: Propaganda, hate speech, or dehumanizing language targeting a group.
Level 3: Glorifying past violence or inciting non-violent illegal acts.
Level 4: Direct incitement to violence or specific threats against a person or group.
Level 5: Immediate threat, planning an attack, or tactical instructions for violence.
"""

#
# 5. PROMPT BUILDER — ZERO-SHOT AND CHAIN-OF-THOUGHT
def build_prompt(text, mode="zero-shot"):
    if mode == "zero-shot":
        system_msg = (
            "You are a content safety expert specializing in extremism assessment. "
            "Your task is to rate the severity of online posts on a scale of 1 to 5. "
            "Do NOT reproduce or amplify the harmful content. Focus only on assessing it.\n\n"
            + SEVERITY_RUBRIC
        )
        user_msg = (
            f"Post: '{text}'\n\n"
            "Based on the severity rubric above, assign a severity level.\n\n"
            "You MUST respond in exactly this format:\n"
            "SEVERITY: [single digit 1-5]\n"
            "EXPLANATION: [your reasoning in 2-4 sentences]"
        )
    else:  # chain-of-thought
        system_msg = (
            "You are a content safety expert specializing in extremism assessment. "
            "Your task is to rate the severity of online posts on a scale of 1 to 5. "
            "Think through your reasoning step by step before giving a final answer. "
            "Do NOT reproduce or amplify the harmful content. Focus only on assessing it.\n\n"
            + SEVERITY_RUBRIC
        )
        user_msg = (
            f"Post: '{text}'\n\n"
            "Work through the following steps to determine severity:\n"
            "Step 1 — Who or what is being targeted?\n"
            "Step 2 — Is there a direct call to violence or just hateful language?\n"
            "Step 3 — How immediate or actionable is any threat?\n"
            "Step 4 — Which severity level (1-5) best fits the rubric?\n\n"
            "You MUST respond in exactly this format:\n"
            "SEVERITY: [single digit 1-5]\n"
            "EXPLANATION: [your step-by-step reasoning in 3-5 sentences]"
        )
    return system_msg, user_msg

# 6. API CALL WITH VISIBLE ERROR HANDLING
def get_severity_assessment(text, mode="zero-shot", temp=0.7):
    system_msg, user_msg = build_prompt(text, mode)
    try:
        response = client.chat.completions.create(
            model=MODEL_B,
            messages=[
                {"role": "system", "content": system_msg},
                {"role": "user",   "content": user_msg}
            ],
            temperature=temp,
            max_tokens=500
        )
        content = response.choices[0].message.content.strip()

        # Parse SEVERITY
        sev_match = re.search(r"SEVERITY:\s*([1-5])", content)
        severity  = sev_match.group(1) if sev_match else "N/A"

        # Parse EXPLANATION (everything after EXPLANATION:)
        exp_match   = re.search(r"EXPLANATION:\s*(.+)", content, re.DOTALL)
        explanation = exp_match.group(1).strip() if exp_match else content

        if severity == "N/A":
            print(f"    Could not parse SEVERITY from response. Raw: {content[:100]}")

        return severity, explanation

    except Exception as e:
        print(f"   API Error [{mode}, temp={temp}]: {e}")
        return "N/A", f"API Error: {e}"

# 7. RUN THE 2x2 EXPERIMENT
results = []

print(f"\nRunning 2×2 experiment: Zero-Shot & CoT × Default temp & 0.4 temp")
print(f"Total API calls: {len(df_part_b) * 4} ({len(df_part_b)} posts × 4 conditions)\n")

for idx, row in tqdm(df_part_b.iterrows(), total=len(df_part_b)):
    text = row['Original_Message']

    # Condition 1: Zero-Shot, Default Temperature (0.7)
    s1, e1 = get_severity_assessment(text, mode="zero-shot", temp=0.7)
    # Condition 2: Zero-Shot, Temperature = 0.4
    s2, e2 = get_severity_assessment(text, mode="zero-shot", temp=0.4)
    # Condition 3: Chain-of-Thought, Default Temperature (0.7)
    s3, e3 = get_severity_assessment(text, mode="cot",       temp=0.7)
    # Condition 4: Chain-of-Thought, Temperature = 0.4
    s4, e4 = get_severity_assessment(text, mode="cot",       temp=0.4)

    results.append({
        "id":         row['id'],
        "text":       text,
        "VE_or_NVE":  row['VE_NVE_classification'],
        "S_Zero_Def": s1, "E_Zero_Def": e1,
        "S_Zero_04":  s2, "E_Zero_04":  e2,
        "S_CoT_Def":  s3, "E_CoT_Def":  e3,
        "S_CoT_04":   s4, "E_CoT_04":   e4,
    })
    time.sleep(0.3)

df_results_b = pd.DataFrame(results)

# 8. SANITY CHECK BEFORE SAVING
print("\n--- Sanity Check ---")
for col in ['S_Zero_Def', 'S_Zero_04', 'S_CoT_Def', 'S_CoT_04']:
    valid = (df_results_b[col] != 'N/A').sum()
    print(f"  {col}: {valid}/100 valid severity scores")

print("\nSample row (first post):")
print(df_results_b.iloc[0][['text','S_Zero_Def','S_Zero_04','S_CoT_Def','S_CoT_04']].to_string())

# 9. SAVE
df_results_b.to_csv('severity_comparison_results.csv', index=False)
print(f"\n✓ Saved severity_comparison_results.csv ({len(df_results_b)} rows)")
```

```python
# 7. LLM COMPARISON: VERBOSITY, READABILITY, AND CLARITY

import textstat

df_all = pd.read_csv('counter_narratives_all.csv')
df_valid = df_all[~df_all['counter_narrative'].str.startswith('API Error', na=False)].copy()

# --- VERBOSITY ---
df_valid['word_count'] = df_valid['counter_narrative'].apply(lambda x: len(str(x).split()))

print("=" * 60)
print("VERBOSITY ANALYSIS")
print("=" * 60)

verbosity = df_valid.groupby('model')['word_count'].agg(['mean', 'median', 'std', 'count'])
verbosity = verbosity.sort_values('mean', ascending=False)
print("\nAverage word count by model:")
print(verbosity.to_string())

most_verbose = verbosity.index[0]
least_verbose = verbosity.index[-1]
print(f"\nMost verbose:  {most_verbose} ({verbosity.loc[most_verbose, 'mean']:.1f} words avg)")
print(f"Least verbose: {least_verbose} ({verbosity.loc[least_verbose, 'mean']:.1f} words avg)")

# --- CONVENTIONAL READABILITY (Flesch Reading Ease & Flesch-Kincaid Grade) ---
print(f"\n{'='*60}")
print("READABILITY ANALYSIS (Conventional Metrics)")
print(f"{'='*60}")

def compute_readability(text):
    text = str(text)
    if len(text.split()) < 5:
        return np.nan, np.nan
    fre = textstat.flesch_reading_ease(text)
    fkg = textstat.flesch_kincaid_grade(text)
    return fre, fkg

df_valid[['flesch_reading_ease', 'flesch_kincaid_grade']] = df_valid['counter_narrative'].apply(
    lambda x: pd.Series(compute_readability(x))
)

def flesch_to_scale(flesch_score):
    if pd.isna(flesch_score):
        return np.nan
    if flesch_score >= 80:
        return 5  # Very easy (5th-6th grade)
    elif flesch_score >= 70:
        return 4  # Easy (7th-8th grade)
    elif flesch_score >= 60:
        return 3  # Standard (9th-10th grade)
    elif flesch_score >= 50:
        return 2  # Somewhat difficult (college)
    else:
        return 1  # Difficult (college graduate)

df_valid['flesch_scale'] = df_valid['flesch_reading_ease'].apply(flesch_to_scale)

readability = df_valid.groupby('model').agg(
    avg_flesch_ease=('flesch_reading_ease', 'mean'),
    avg_fk_grade=('flesch_kincaid_grade', 'mean'),
    avg_flesch_scale=('flesch_scale', 'mean')
).round(2)

print("\nReadability by model:")
print(readability.to_string())

# --- LLM-BASED CLARITY SCORING (using gpt-4.1-mini) ---
print(f"\n{'='*60}")
print("LLM CLARITY SCORING (GPT-4.1-mini as judge)")
print(f"{'='*60}")

def get_llm_clarity_score(counter_narrative):
    try:
        response = client.chat.completions.create(
            model="gpt-4.1-mini",
            messages=[
                {"role": "system", "content": (
                    "You are an expert in evaluating text clarity for a general public audience "
                    "(8th grade reading level). Rate the following counter-narrative on a scale of 1-5:\n"
                    "1 = Very unclear, confusing, jargon-heavy\n"
                    "2 = Somewhat unclear, hard to follow\n"
                    "3 = Moderately clear, understandable with effort\n"
                    "4 = Clear, easy to understand\n"
                    "5 = Very clear, immediately understandable\n\n"
                    "Respond with ONLY a single digit (1-5)."
                )},
                {"role": "user", "content": f"Counter-narrative:\n\"{counter_narrative}\""}
            ],
            temperature=0.0,
            max_tokens=5
        )
        score = re.search(r'[1-5]', response.choices[0].message.content)
        return int(score.group()) if score else np.nan
    except Exception as e:
        print(f"  Clarity scoring error: {e}")
        return np.nan

# Score a representative sample (one per model per persona, ~16 examples)
sample_for_scoring = df_valid.groupby(['model', 'persona']).first().reset_index()
print(f"Scoring {len(sample_for_scoring)} counter-narratives for LLM clarity...")

clarity_scores = []
for idx, row in tqdm(sample_for_scoring.iterrows(), total=len(sample_for_scoring)):
    score = get_llm_clarity_score(row['counter_narrative'])
    clarity_scores.append(score)
    time.sleep(0.3)

sample_for_scoring['llm_clarity_score'] = clarity_scores

# Merge back
df_valid = df_valid.merge(
    sample_for_scoring[['model', 'persona', 'llm_clarity_score']],
    on=['model', 'persona'], how='left', suffixes=('', '_sample')
)

clarity_by_model = sample_for_scoring.groupby('model')['llm_clarity_score'].mean().round(2)
print("\nLLM Clarity scores by model:")
print(clarity_by_model.to_string())

# 8. COMPARISON ANALYSIS: CONVENTIONAL vs LLM READABILITY
print(f"\n{'='*60}")
print("COMPARISON: CONVENTIONAL vs LLM READABILITY")
print(f"{'='*60}")

comparison_df = sample_for_scoring[['model', 'persona_name', 'word_count',
                                     'flesch_scale', 'llm_clarity_score',
                                     'counter_narrative']].copy()
comparison_df['disagree'] = comparison_df['flesch_scale'] != comparison_df['llm_clarity_score']

disagree_examples = comparison_df[comparison_df['disagree']].head(5)
print(f"\nDisagreement rate: {comparison_df['disagree'].mean()*100:.1f}%")
print(f"\n5 Examples where conventional and LLM scores disagree:")
print("-" * 60)
for i, row in disagree_examples.iterrows():
    text_preview = row['counter_narrative'][:100] + "..." if len(str(row['counter_narrative'])) > 100 else row['counter_narrative']
    print(f"\n  Model: {row['model']} | Persona: {row['persona_name']}")
    print(f"  Flesch scale: {row['flesch_scale']} | LLM clarity: {row['llm_clarity_score']}")
    print(f"  Text: {text_preview}")

print(f"\n{'='*60}")
print("ANALYSIS SUMMARY")
print(f"{'='*60}")
print("""
Which metric is more accurate for counter-narratives?
The LLM-based clarity score is likely more accurate for counter-narratives because:
  1. Flesch metrics are formula-based (syllable/word/sentence counts) and can misjudge
     texts that use simple words in complex argumentation structures.
  2. Counter-narratives require nuanced reasoning. A text can have simple vocabulary
     (high Flesch score) but still be logically hard to follow.
  3. The LLM evaluates semantic clarity and coherence, not just surface-level readability.

Relationship between verbosity and readability:
More verbose counter-narratives tend to have lower Flesch Reading Ease scores (harder to read)
  because longer texts often contain more complex sentence structures. However, LLM clarity
  scores do not always penalize length. A longer, well-structured response can still be rated
  as highly clear. This suggests verbosity and readability have a negative correlation at the
  surface level (Flesch), but a more nuanced relationship when judged by LLM.
""")

# Save final analysis
df_valid.to_csv('counter_narratives_analyzed.csv', index=False)
print("Saved: counter_narratives_analyzed.csv")
```
